## Visualization & Explanatory Analysis


### Stasiun mana yang tercatat memiliki PM2.5 dan PM10 tertinggi?


```python
max_pm25 = df.groupby(by="station").agg({
    "PM2.5": "max",
}).sort_values(by="PM2.5", ascending=False).head(5)

max_pm10 = df.groupby(by="station").agg({
    "PM10": "max",
}).sort_values(by="PM10", ascending=False).head(5)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 8))

sns.barplot(x="PM2.5", y="station", data=max_pm25, hue="station",
            palette="colorblind", ax=ax[0])
ax[0].set_xlabel(None)
ax[0].set_ylabel(None)
ax[0].set_title("Stasiun Polusi PM2.5 Terbanyak", loc="center", fontsize=18)
ax[0].tick_params(axis='y', labelsize=12)


sns.barplot(x="PM10", y="station", data=max_pm10, hue="station",
            palette="colorblind", ax=ax[1])
ax[1].set_xlabel(None)
ax[1].set_ylabel(None)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Stasiun Polusi PM10 Terbanyak", loc="center", fontsize=18)
ax[1].tick_params(axis='y', labelsize=12)

plt.suptitle("Stasiun Polusi PM2.5 dan PM10 Terbanyak", fontsize=32)
```

Dari grafik tersebut dapat disimpulkan bahwa stasiun dengan konsentrasi PM2.5 tertinggi adalah Wanshouxigong dan stasiun dengan konsentrasi PM10 tertinggi adalah Changping.


### Bagaimana korelasi PM2.5 dan PM10 dengan suhu?


```python
correlation = df[['PM2.5', 'PM10', 'TEMP']].corr()
```

```python
sns.heatmap(data=correlation,
            annot=True, fmt=".3f", linewidths=.5)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))

sns.scatterplot(x=df['TEMP'], y=df['PM2.5'], ax=ax[0])
ax[0].set_xlabel("Temperature")
ax[0].set_ylabel("PM2.5")
ax[0].set_title("Pengaruh Temperatur terhadap PM2.5",
                loc="center", fontsize=18)
ax[0].tick_params(axis='y', labelsize=12)

sns.scatterplot(x=df['TEMP'], y=df['PM10'], ax=ax[1])
ax[1].set_xlabel("Temperature")
ax[1].set_ylabel("PM10")
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Pengaruh Temperatur terhadap PM10",
                loc="center", fontsize=18)
ax[1].tick_params(axis='y', labelsize=12)

plt.show()
```

Suhu memiliki korelasi negatif yang lemah dengan PM2.5 dan PM10, yang berarti bahwa peningkatan suhu mungkin sedikit mengurangi konsentrasi kedua polutan tersebut, tetapi hubungan ini tidak signifikan secara statistik.


### Apakah kecepatan angin mempengaruhi kadar PM2.5 dan PM10?


```python
correlation = df[['PM2.5', 'PM10', 'WSPM']].corr()
```

```python
sns.heatmap(data=correlation,
            annot=True, fmt=".3f", linewidths=.5)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))

sns.scatterplot(x=df['WSPM'], y=df['PM2.5'], ax=ax[0])
ax[0].set_xlabel("Wind Speed")
ax[0].set_ylabel("PM2.5")
ax[0].set_title("Pengaruh Kecepatan Angin terhadap PM2.5",
                loc="center", fontsize=18)
ax[0].tick_params(axis='y', labelsize=12)

sns.scatterplot(x=df['WSPM'], y=df['PM10'], ax=ax[1])
ax[1].set_xlabel("Wind Speed")
ax[1].set_ylabel("PM10")
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Pengaruh Kecepatan Angin terhadap PM10",
                loc="center", fontsize=18)
ax[1].tick_params(axis='y', labelsize=12)

plt.show()
```

Kecepatan angin memiliki korelasi negatif dengan PM2.5 dan PM10, yang berarti bahwa peningkatan kecepatan angin cenderung mengurangi konsentrasi kedua polutan tersebut.


Kecepatan angin (WSPM) lebih tinggi berkorelasi dengan penurunan konsentrasi polutan, karena angin membantu penyebaran polutan di udara.


### Bagaimana perubahan suhu di tiap stasiun setiap tahun?


```python
df['date'] = pd.to_datetime(df[['year', 'month', 'day']])

monthly_avg_temp = df.groupby([df['station'], df['date'].dt.to_period('M')])[
    'TEMP'].mean().reset_index()

monthly_avg_temp['date'] = monthly_avg_temp['date'].dt.to_timestamp()
```

```python
plt.figure(figsize=(12, 8))

sns.lineplot(data=monthly_avg_temp, x='date', y='TEMP',
             hue='station', marker='o')

plt.title('Temperature Change by Station')
plt.xlabel(None)
plt.ylabel('Average Temperature')
plt.legend(title='Station')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Dari grafik dapat dilihat bahwa terdapat perubahan suhu pada tiap musim yang jelas dan perbedaan suhu tiap stasiun tidak jauh berbeda.
